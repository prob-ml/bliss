# Sanity check

```python
ds = galaxy_datasets.CatsimGalaxies(image_size=40, fixed_size=False)
```

```python
import random
random_idx = random.randrange(len(ds))
img = ds[random_idx]['image']
back = ds[random_idx]['background']
img1 = img[0]
back1 = back[0]

# is the var >= mean in a dark region
plt.imshow(img1[0:20, 0:10])
print(np.mean(img1[0:5, 0:5]))
print(np.var(img1[0:5,0:5]))
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable


fig, axs = plt.subplots(1,3,figsize=(14,14))
ax1,ax2,ax3 = axs.flatten()


im = ax1.imshow(img1)
divider = make_axes_locatable(ax1)
cax = divider.append_axes('right', size='5%', pad=0.05)
fig.colorbar(im, cax=cax, orientation='vertical')


im = ax2.imshow(img1-back1)
divider = make_axes_locatable(ax2)
cax = divider.append_axes('right', size='3%', pad=0.05)
fig.colorbar(im, cax=cax, orientation='vertical')

im = ax3.imshow(back1)
divider = make_axes_locatable(ax3)
cax = divider.append_axes('right', size='5%', pad=0.05)
fig.colorbar(im, cax=cax, orientation='vertical')

plt.tight_layout()
```
