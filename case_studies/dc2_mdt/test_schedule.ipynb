```python
import torch
import matplotlib.pyplot as plt
```

```python
sigma = torch.linspace(0, 1, steps=1000)
alpha = 1.0 - sigma
bs = [-2, -1, 0, 1, 2]
eps = 1e-3
```

```python
alpha_2 = alpha ** 2 + eps
sigma_2 = sigma ** 2 + eps
loss_weights = {b: 1 / (1 + torch.exp(b - torch.log(alpha_2 / sigma_2)))
                for b in bs}
```

```python
for k, v in loss_weights.items():
    plt.plot(list(range(1000)), v, label=f"b={k}")
plt.xlabel("Time Step")
plt.ylabel("Loss Weight")
plt.legend()
plt.show()
```
