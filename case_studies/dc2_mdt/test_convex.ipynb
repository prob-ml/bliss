```python
import torch
import math
import matplotlib.pyplot as plt
```

```python
ir = []
part1 = 10
part2 = 1000
for i in range(part1):
    jr = []
    basis_fn_angle = math.pi * i / part1
    basis_function = torch.tensor([[1.0, 0.0],
                                  [math.cos(basis_fn_angle), math.sin(basis_fn_angle)],
                                  [math.cos(basis_fn_angle + math.pi / 4), math.sin(basis_fn_angle + math.pi / 4)],
                                  [math.cos(basis_fn_angle + math.pi / 3), math.sin(basis_fn_angle + math.pi / 3)]])
    for j in range(part2):
        coeff_angle = 2 * math.pi * j / part2
        coeff = torch.tensor([math.cos(coeff_angle), math.sin(coeff_angle)]).view(2, 1)
        num = (basis_function[0] @ coeff * 1).exp()
        denom = (basis_function @ coeff * 1).exp().sum()
        jr.append(num / denom)
    jr = torch.cat(jr, dim=0)
    ir.append(jr)
ir = torch.stack(ir, dim=0)
```

```python
x = torch.linspace(0, 2 * torch.pi, part2)
for i, r in enumerate(ir):
    plt.figure(figsize=(5, 5))
    plt.plot(x, r, label=f"{i}/{part1} * pi")
    plt.legend()
    plt.show()
```
