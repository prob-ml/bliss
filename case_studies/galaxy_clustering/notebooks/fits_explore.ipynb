```python
import numpy as np

# Set up matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

from astropy.io import fits
from matplotlib.colors import LogNorm
```

```python
image_r = fits.getdata('/home/kapnadak/bliss/case_studies/galaxy_clustering//data/images/galsim_des_001_r.fits')
image_g = fits.getdata('/home/kapnadak/bliss/case_studies/galaxy_clustering//data/images/galsim_des_001_g.fits')
image_i = fits.getdata('/home/kapnadak/bliss/case_studies/galaxy_clustering//data/images/galsim_des_001_i.fits')
image_z = fits.getdata('/home/kapnadak/bliss/case_studies/galaxy_clustering//data/images/galsim_des_001_z.fits')
```

```python
plt.imshow(image_r, norm=LogNorm())
plt.colorbar()
```

```python
from mpl_toolkits.axes_grid1 import ImageGrid
fig = plt.figure(figsize=(12., 12.))
grid = ImageGrid(fig, 111,  # similar to subplot(111)
                 nrows_ncols=(2, 2),  # creates 2x2 grid of axes
                 axes_pad=0.1,  # pad between axes in inch.
                 )

for ax, im in zip(grid, [image_g, image_r, image_i, image_z]):
    ax.imshow(im, norm = LogNorm())
```

```python
image_overall = np.stack([image_i, image_r, image_g], axis=2)
image_overall.shape
```
