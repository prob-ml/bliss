```python
import case_studies.galaxy_clustering.data_generation.DES_data_extraction as des_extraction
from matplotlib import pyplot as plt
from matplotlib.colors import LogNorm
from astropy.io import fits
import os
import pandas as pd
```

```python
columns = [
    "Cluster_ID",
    "Cluster_Name",
    "RA",
    "DEC",
    "z_cl",
    "BCG_i",
    "BCG_W1",
    "SNR",
    "r_500",
    "richness",
    "M_500",
    "N_500",
    "references"
]
catalog = pd.read_fwf("/home/kapnadak/bliss/case_studies/galaxy_clustering/cluster_DESunWISE.dat", names=columns)
```

```python
mean_mass, std_mass = catalog["M_500"].mean(), catalog["M_500"].std()
mean_z, std_z = catalog["z_cl"].mean(), catalog["z_cl"].std()
mean_r, std_r = catalog["r_500"].mean(), catalog["r_500"].std()
mean_n, std_n = catalog["N_500"].mean(), catalog["N_500"].std()
print("Summary Statistics:")
print(f"Mean M500: {mean_mass}, Std M500: {std_mass}")
print(f"Mean r500: {mean_r}, Std r500: {std_r}")
print(f"Mean N500: {mean_n}, Std N500: {std_n}")
print(f"Mean z: {mean_z}, Std z: {std_z}")
```

```python
catalog.sample(n=1)
```
