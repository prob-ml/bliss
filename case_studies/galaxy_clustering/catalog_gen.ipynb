```python
from prior import ClusterPrior
from astropy.io import ascii
from astropy.table import Table
cluster_prior_obj = ClusterPrior()
catalogs = cluster_prior_obj.sample()
catalogs[0]
```

```python
for i in range(len(catalogs)):
    catalog = catalogs[i]
    file_name = f"catalogs/catalog_{i:03}.dat"
    catalog_table = Table.from_pandas(catalog)
    ascii.write(catalog_table, file_name, format="no_header", overwrite=True)
```
